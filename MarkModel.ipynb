```python
%matplotlib inline
import pandas
from keras.layers import Input
df = pandas.read_csv('Marks-Dataset.csv')
```

```python
X = df[['Visits', 'Homework', 'Test']].values
Y = df['Exam'].values
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
model = Sequential()
model.add(Dense(1, input_shape=(3,)))
```

```python
model.summary()
```

```python
model.compile(Adam(lr=0.8), 'mean_squared_error')
```

```python
model.fit(X, Y, epochs=100)
```

```python
import coremltools
coreml_model = coremltools.converters.keras.convert(model, input_names='matrixVisitsHomeworkTest', output_names='examMark')
coreml_model.save('MarkModel.mlmodel')
```
